```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import openpyxl
import pandas as pd
import matplotlib.dates as mdates

fig = plt.figure(figsize=(50, 20))
ax1 = fig.subplots()
ax2 = ax1.twinx()

df_2012 = pd.read_csv("2012.csv", header=1)
df_2013 = pd.read_csv("2013.csv", header=1)
df_2014 = pd.read_csv("2014.csv", header=1)
df_2015 = pd.read_csv("2015.csv", header=1)
df_2016 = pd.read_csv("2016.csv", header=1)
df_2017 = pd.read_csv("2017.csv", header=1)
df_2018 = pd.read_csv("2018.csv", header=1)
df_2019 = pd.read_csv("2019.csv", header=1)
df_2020 = pd.read_csv("2020.csv", header=1)
df_2021 = pd.read_csv("2021.csv", header=1)

df = pd.concat([df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021])
df.head()
#df['day']=pd.to_timedelta(df[''],unit='D')+pd.to_datetime("1899/12/30") 
a = df[''].values
b = df[''].values
c = df[''].values
ax1.plot(a,b,color='r',label='1-0',marker="*",linestyle="None")
ax2.plot(a,c,color='g',label='%')
ax1.set_ylim([0,1])
ax2.set_ylim([0,100])
ax1.set_ylabel(df[''].name)
ax2.set_ylabel(df[''].name)
plt.grid(True)
#plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=1))
#x_dt = pd.to_datetime(df[''])
fig.autofmt_xdate(rotation=45)
plt.savefig("fig1.png")
plt.show()
```
