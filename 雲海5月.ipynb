```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import openpyxl
import pandas as pd
import matplotlib.dates as mdates

fig = plt.figure(figsize=(50, 20))
ax1 = fig.subplots()
ax2 = ax1.twinx()

df_5 = pd.read_csv("雲海5月.csv")
df = df_5
df.head()
a = df['日付'].values
b = df['出現'].values
c = df['確率'].values
ax1.plot(a,b,color='r',label='1-0',marker="*",linestyle="None")
ax2.plot(a,c,color='g',label='%')
ax1.set_ylim([0,1])
ax2.set_ylim([0,100])
ax1.set_ylabel(df['出現'].name)
ax2.set_ylabel(df['確率'].name)
plt.grid(True)
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=1))
fig.autofmt_xdate(rotation=45)
plt.savefig("fig1.png")
plt.show()
```
